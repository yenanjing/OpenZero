```python
import re


def format_answer(row):
    res = {"first_industry": "", "second_industry": ""}
    answer = row["正确答案"].strip()
    if row["答案来源"] == "人工":
        res["first_industry"] = answer.split("-")[0].strip()
        res["second_industry"] = answer.split("-")[1].strip()
    elif row["答案来源"] == "混元turbo":
        try:
            pattern = r'(?:一级行业“(.*?)”|“(.*?)”一级行业)'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["first_industry"] = ind.strip()
            pattern = r'的“(.*?)”二级行业'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["second_industry"] = ind.strip()
        except:
            print(answer)
    elif row["答案来源"] == "deepseek-R1":
        try:
            pattern = r'一级行业：?“?(.*?)”?(?:，|\n|。|；)'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["first_industry"] = ind.strip()
            pattern = r'；?二级行业：?“?(.*?)”?(?:，|\n|。)'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["second_industry"] = ind.strip()
        except:
            print(answer)
    if res["first_industry"] == "" or res["second_industry"] == "":
        print("error", row["问题描述"])
    return res


df["answer"] = df.apply(format_answer, axis=1)
df
```

# 混元模型测试

```python
import uuid
# 调用混元api
import requests
import time
import random
import numpy as np
from concurrent.futures import ThreadPoolExecutor

yaml_file_path = os.path.join(current_dir, "prompt_hy.yaml")
with open(yaml_file_path, "r", encoding="utf-8") as file:
    prompt_data = yaml.safe_load(file)

prompt_t = prompt_data["prompt"]
wsid = "10419"

headers = {
    'Authorization': "Bearer 7auGXNATFSKl7dF",
    'Content-Type': 'application/json',
    "Wsid": wsid,
}
model = "kefu_open_industry_train_gen_test"
ss_url = "http://stream-server-online-openapi.turbotke.production.polaris:8080/openapi/chat/completions"


def request_hy(prompt):
    prompt = prompt_t.replace("{query}", prompt)
    # print(prompt)
    json_data = {
        'model': model,
        "query_id": "test_query_id_" + str(uuid.uuid4()),
        "messages": [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        "stream": False,
        "max_input_seq_len": 16384,
        "temperature": 0,
        "top_k": 1,
        "top_p": 0,
    }
    response = requests.post(ss_url,
                             headers=headers,
                             json=json_data,
                             stream=False)

    # print(response.text)
    if response.status_code == 200:
        try:
            if response.headers['Content-Type'] == 'text/event-stream':
                # 逐行读取响应内容
                for line in response.iter_lines():
                    if line:  # 确保行不为空
                        print(line)
            elif response.headers['Content-Type'] == 'application/json':
                result = response.json()
                if 'choices' not in result:
                    print(result)
                    return ""
                return result['choices'][0]['message']['content']
        except Exception as e:
            print(response.headers['Content-Type'])
            print("发生异常：", e)
    else:
        print("请求失败，状态码：", response.status_code)
        return None


def concurrent_requests(prompts, num_requests):
    answers = []
    with ThreadPoolExecutor(max_workers=num_requests) as executor:
        futures = {executor.submit(request_hy, prompts[i]): i for i in range(num_requests)}
        for future in futures:
            result = future.result()
            if result is not None:
                answers.append(result)
    return answers


# prompts = ['南极为啥没有北极熊，一句话回答', '介绍林黛玉', '天空网是啥', '写篇父亲节的作文，800字的', 
#            '如何看待中美战争', '德国为啥战败', '写个df删除2列的代码', 'java写个冒泡排序']

# num_requests = len(prompts)  # 设置并发请求数量
# answers = concurrent_requests(prompts, num_requests)
# print(answers)
request_hy("图书租赁开什么行业")
```

```python
import pandas as pd

df_test = pd.read_excel("/Users/wenweili/Desktop/AI客服/开户行业/妙问开户行业数据集_v4_test.xlsx")
df_test
```

```python
results = []
f_c = 0
s_c = 0

for idx, row in df_test.iterrows():
    prompt = row["问题描述"]
    answer = request_hy(prompt)
    gt = eval(row["answer"])
    f_g = gt["first_industry"]
    s_g = gt["second_industry"]
    try:
        f_a = answer.split("；")[0].split("：")[1].strip()
        s_a = answer.split("；")[1].split("：")[1].strip()
    except:
        f_a = ""
        s_a = ""
    results.append([prompt, f_g, s_g, f_a, s_a])
    if f_g == f_a:
        f_c += 1
    if s_g == s_a:
        s_c += 1
    if idx < 5:
        print(row["问题描述"])
        print("gt:", f_g, s_g)
        print("pred:", f_a, s_a)
        print()
    
print("一级行业准确率：" + str(f_c / len(df_test)))
print("二级行业准确率：" + str(s_c / len(df_test)))
```

```python
results_df = pd.DataFrame(results, columns=["prompt", "first_gt", "second_gt", "first_pred", "second_pred"])
results_df
```
