```python
import re

def format_answer(row):
    res = {"first_industry": "", "second_industry": ""}
    answer = row["正确答案"].strip()
    if row["答案来源"] == "人工":
        res["first_industry"] = answer.split("-")[0].strip()
        res["second_industry"] = answer.split("-")[1].strip()
    elif row["答案来源"] == "混元turbo":
        try:
            pattern = r'(?:一级行业“(.*?)”|“(.*?)”一级行业)'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["first_industry"] = ind.strip()
            pattern = r'的“(.*?)”二级行业'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["second_industry"] = ind.strip()
        except:
            print(answer)
    elif row["答案来源"] == "deepseek-R1":
        try:
            pattern = r'一级行业：?“?(.*?)”?(?:，|\n|。)'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["first_industry"] = ind.strip()
            pattern = r'二级行业：?“?(.*?)”?(?:，|\n|。)'
            match = re.search(pattern, answer)
            ind = match.group(1) if match.group(1) else match.group(2)
            res["second_industry"] = ind.strip()
        except:
            print(answer)
    if res["first_industry"] == "" or res["second_industry"] == "":
        print("error", row["问题描述"])
    return res

df["answer"] = df.apply(format_answer, axis=1)
df
```

```python
df_test = df[df["答案来源"].isin(["人工", "混元turbo"])]
df_test = df_test[["问题描述", "answer"]]
df_test
```
